```python
from functools import lru_cache
import sys
sys.setrecursionlimit(10**7)

@lru_cache(None)
def f(a, b , n):
    if a + b >= 47:
        return 0
    
    if a == b:
        t = [f(a + i, b, n + 1) for i in range(1, 4)] + [f(a, b + i, n + 1) for i in range(1, 4)]
    if a > b:
        t = [f(a + i, b, n + 1) for i in range(1, 4)] + [f(a, b * 2, n + 1)]
    else:
        t = [f(a, b + i, n + 1) for i in range(1, 4)] + [f(a * 2, b, n + 1)]

    n = [i for i in t if i <= 0]
    if n:
        return -max(n) + 1
    return -max(t)
    # 19

ans = []
for s1 in range(0, 46):
    for s2 in range(0, 46):
        if f(s1, s2, 0) == 1:
            ans.append(s1 + s2)
print(min(ans))
```

1) убрать из кучи пять камней;
2) если количество камней в куче чётно, уменьшить его в два раза;
3) если количество камней в куче кратно трём, уменьшить его в три раза;

Например, если в куче 12 камней, то за один ход можно получить 7, 6 или 4 камня, а если в куче 11 камней, то за один ход можно получить 6 или 12 камней. Игра завершается, когда количество камней в куче становится не более 19. Победителем считается игрок, сделавший последний ход, то есть первым получивший кучу, в которой будет 19 или меньше камней. Задание 21.
Для игры, описанной в задании 19, найдите минимальное значение S, при котором у Вани есть стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Пети, но у Вани нет стратегии, которая позволила бы ему гарантированно выиграть первым ходом.

```python
from functools import lru_cache

@lru_cache(None)
def f(a, n):
    if a <= 19:
        return 0
    
    t = [f(a - 5, n + 1)]
    if a % 2 == 0:
        t += [f(a // 2, n + 1)]
    if a % 3 == 0:
        t += [f(a // 3, n + 1)]
    if a % 2 and a % 3:
        t += [f(a + 1, n + 1)]

    n = [i for i in t if i <= 0]
    if n:
        return -max(n) + 1
    return -max(t)

ans = []
# 19
for s in range(150, 18, -1):
    if f(s, 0) != 1 and f(s, 0) == -1:
        ans.append(s)
print(min(ans))
ans.clear()

# 20
for s in range(150, 18, -1):
    if f(s, 0) != 1 and f(s, 0) == 2:
        ans.append(s)
print(*ans)
ans.clear()

# 21
for s in range(150 ,18, -1):
    if f(s, 0) == -2:
        ans.append(s)
print(min(ans))
```
