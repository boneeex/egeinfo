Найдите такое значение S, при котором у Вани есть стратегия, позволяющая ему выиграть вторым ходом при любой игре Пети, но у Вани нет стратегии, которая позволяла бы ему гарантированно выиграть первым ходом.

```python
from functools import lru_cache
import sys
sys.setrecursionlimit(10**7)

@lru_cache(None)
def f(a, b , n):
    if a + b >= 47:
        return 0
    
    if a == b:
        t = [f(a + i, b, n + 1) for i in range(1, 4)] + [f(a, b + i, n + 1) for i in range(1, 4)]
    if a > b:
        t = [f(a + i, b, n + 1) for i in range(1, 4)] + [f(a, b * 2, n + 1)]
    else:
        t = [f(a, b + i, n + 1) for i in range(1, 4)] + [f(a * 2, b, n + 1)]

    n = [i for i in t if i <= 0]
    if n:
        return -max(n) + 1
    return -max(t)
    # 19

ans = []
for s1 in range(0, 46):
    for s2 in range(0, 46):
        if f(s1, s2, 0) == 1:
            ans.append(s1 + s2)
print(min(ans))
```
