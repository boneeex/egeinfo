Определите максимальное количество символов в непрерывной последовательности, чьё значение выражения является чётным, которая является корректным арифметическим выражением с целыми неотрицательными числами, записанными в десятичной системе счисления.

При этом искомая последовательность не должна являться частью другой, более длинной непрерывной последовательности, которая также является корректным арифметическим выражением.

В этом выражении никакие два знака арифметических операций не стоят рядом, в записи чисел отсутствуют незначащие (ведущие) нули.

```python
from re import *
file = open(r"D:\informaticsclass\egeinfo\24\24_1__6ba3v.txt").readline()

num = "([1-9][0-9]*|0)"
reg = f"{num}([+*]{num})+"
mx = 0

for i in finditer(reg, file):
    a = i.group()
    if eval(a) % 2 == 0:
        mx = max(mx, len(a))

print(mx)
```

Текстовый файл 24_7.txt состоит из символов, обозначающих знаки «+», «*» и десятичных цифр.

Определите максимальное значение, которое является результатом вычисления непрерывной последовательности, являющейся корректным арифметическим выражением с целыми неотрицательными числами, в котором обязательно присутствует операция умножения, но операция сложения может как присутствовать, так и полностью отсутствовать.

При этом искомая последовательность не должна являться частью другой, более длинной непрерывной последовательности, которая также является корректным арифметическим выражением.

В этом выражении никакие два знака арифметических операций не стоят рядом. В записи чисел отсутствуют незначащие (ведущие) нули и число 0 не имеет знака.

В ответе укажите значение полученного выражения.

```python
from re import *
file = open(r"D:\informaticsclass\egeinfo\24\24_7__6ao2p.txt").readline()
mx = 0

num = "([1-9][0-9]*|0)"
reg = f"{num}([+*]{num})+"

for i in finditer(reg, file):
    if str(i.group()).count("*") >= 1:
        mx = max(mx, eval(i.group()))
print(mx)
```
