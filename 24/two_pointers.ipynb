Текстовый файл содержит строку из заглавных латинских букв и точек, всего не более 106 символов. Определите максимальное количество идущих подряд символов, среди которых не более шести точек и более 15 гласных букв (AEIOUY).

```python
from collections import defaultdict
file = open(r"D:\informaticsclass\egeinfo\24\24_M3__42nfh.txt")
file = file.readline()

d = defaultdict(int)
p = mx = start = 0
for end in range(len(file)):
    if file[end] in "AEIOUY":
        d[file[end]] += 1
    if file[end] == ".":
        p += 1

    while p > 6:
        if file[start] in "AEIOUY":
            d[file[end]] -= 1
        if file[start] == ".":
            p -= 1
        start += 1
    if p <= 6 and sum(d.values()) > 15:
        mx = max(mx, end - start + 1)
print(mx)
```

Текстовый файл состоит из символов T,U,V,W,X,Y и Z.
Определите в прилагаемом файле максимальное количество идущих подряд символов (длину непрерывной подпоследовательности), среди которых символ T   встречается ровно 100 раз.
Для выполнения этого задания следует написать программу

```python
file = open(r"D:\informaticsclass\egeinfo\24\24_2024__7aiiq.txt")
file = file.readline()

mx = t = start = 0
for end in range(len(file)):
    if file[end] == "T":
        t += 1
    while t > 100:
        if file[start] == "T":
            t -= 1
        start += 1
    if t == 100:
        mx = max(mx, end - start + 1)

print(mx)
```

Текстовый файл содержит только заглавные буквы латинского алфавита. Определите максимальное количество идущих подряд символов, 
среди которых каждая из букв F, L встречается не более трех раз.

```python
file = open(r"D:\informaticsclass\egeinfo\24\24_18__3b9tx.txt").readline()

mx = f = l = start = 0
for end in range(len(file)):
    if file[end] == "F":
        f += 1
    if file[end] == "L":
        l += 1
    
    while f > 3 or l > 3:
        if file[start] == "F":
            f -= 1
        if file[start] == "L":
            l -= 1
        start += 1
    if f <= 3 and l <= 3:
        mx = max(mx, end - start + 1)
print(mx)
```

Текстовый файл состоит не более чем из 106 символов латинского алфавита. Определите длину минимальной подстроки, содержащей не менее 100 символов Y.

Для выполнения этого задания следует написать программу.

```python
f = open(r"D:\informaticsclass\egeinfo\24\24_16__3b9u2.txt").readline()

start = y = 0
mn = 10 ** 7

for end in range(len(f)):
    if f[end] == "Y":
        y += 1

    while y > 100:
        if f[start] == "Y":
            y -= 1
        start += 1
    
    if y == 100:
        mn = min(mn, end - start + 1)

print(mn)
```
