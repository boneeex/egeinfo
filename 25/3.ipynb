Найдите все натуральные числа, принадлежащие отрезку [113 000 000; 114 000 000], у которых ровно три различных чётных делителя.
В ответе перечислите найденные числа в порядке возрастания, справа от каждого числа запишите его второй по величине чётный делитель.

```python
from math import *

def prost(n):
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

ans = []

for i in range(113000000, 114000000 + 1, 2):
    q = sqrt(i // 2)
    if q == int(q) and prost(q):
        ans.append((i, 2 * q))

for i in ans:
    print(i)
```

Найдите все натуральные числа, принадлежащие отрезку [55 000 000; 60 000 000], у которых ровно пять различных нечётных делителей (количество чётных делителей может быть любым). В ответе перечислите найденные числа, справа от каждого числа запишите его наибольший нечётный делитель.

```python
def del_two(n):
    while n % 2 == 0:
        n //= 2
    return n

from math import *

def prost(n):
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


ans = []
for x in range(55000000, 60000000):
    x1 = del_two(x)
    x1sq = sqrt(sqrt(x1))
    if x1sq == int(x1sq) and prost(x1sq):
        ans.append((x, x1))

for t in ans:
    print(t)
```
